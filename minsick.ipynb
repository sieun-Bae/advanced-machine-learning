```python
data = open("spambase/spambase.data")
data = str(data.read()) #string으로 변환
```

```python
data = data.split('\n') #\n 기준으로 data split
```

```python
data
```

```python
len(data) #전체 엔트리 수
```

```python
l_data = list()
for d in data:
    l_data.append(d.split(',')) #, 기준 split
```

```python
import pandas as pd
df = pd.DataFrame(l_data) #dataframe 으로 변환
```

```python
df.head()
```

```python
df.info()
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```
